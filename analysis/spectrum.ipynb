### load data file and format array

```python
stridat = "./092509_vna_coil current (mA)mw frequency (Hz).dat"
data = np.genfromtxt(stridat).T
```

```python
row_x = 0
row_y = 1
row_d = 6
```

```python
x_val = np.unique(data[row_x])
len(x_val)
```

```python
y_val = np.extract(data[row_x] == data[row_x][0],data[row_y])
len(y_val)
```

```python
m_dat_arr = []
for y in y_val:
    m_dat_arr.append(np.extract(data[row_y] == y,data[row_d]))
```

```python
m_dat = np.array(m_dat_arr)
m_dat.shape
```

### Normalization and masking

```python
def norm_mask(tolerance=0.5):
    m = m_dat.T
    mask_a = []
    for c in range(m.shape[0]):
        lower_th = np.mean(m[c][np.nonzero(m[c])]) - tolerance*np.abs(np.mean(m[c][np.nonzero(m[c])]))
        upper_th = np.mean(m[c][np.nonzero(m[c])]) + tolerance*np.abs(np.mean(m[c][np.nonzero(m[c])]))
        mask = np.any((m[c]<lower_th,m[c]>upper_th,m[c]==0),axis=0)
        mask_a.append(mask)
        #print mask
        min_v =  np.ma.min(np.ma.masked_array(m[c], mask))
        pp = np.ma.max(np.ma.masked_array(m[c], mask)) - min_v
        m[c] = (m[c] - min_v) / pp
    return m, mask_a
```

```python
m, mask_a = norm_mask()
```

#### find maxima

```python
f01 = []
for i in range(len(x_val)):
    f01.append(np.extract(m[i] == np.ma.min(np.ma.masked_array(m[i], mask_a[i])),y_val))
```

```python
h = 6.626e-34
e = 1.6022e-19
phi0 = h/(2*e)
Ec = 0.1157
d = 0.1
```

```python
def f_transmon_disp(I, Ej, dphi, a):
    #return np.sqrt(8*Ec*Ej*np.abs(np.cos(np.pi*a*I+dphi)))
    argument = np.pi*a*I+dphi
    return np.sqrt(8*Ec*Ej*np.abs(np.cos(argument))*np.sqrt(np.abs(1+(d*np.tan(argument))**2)))-Ec
```

```python
popt, pcov = curve_fit(f_transmon_disp, x_val, 1e-9*np.array(f01).squeeze(), p0=[70,2.3,0.31])
```

```python
Ic_tot = 2*h*np.pi*1e9*popt[0]/phi0
print 'Ic = '+str(round(1e9*Ic_tot,3))+'nA'
```
