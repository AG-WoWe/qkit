### Normalization and masking

```python
def norm_mask(m, tolerance=0.5):
    #m = m_dat.T
    mask_a = []
    for c in range(m.shape[0]):
        lower_th = np.mean(m[c][np.nonzero(m[c])]) - tolerance*np.abs(np.mean(m[c][np.nonzero(m[c])]))
        upper_th = np.mean(m[c][np.nonzero(m[c])]) + tolerance*np.abs(np.mean(m[c][np.nonzero(m[c])]))
        mask = np.any((m[c]<lower_th,m[c]>upper_th,m[c]==0),axis=0)
        mask_a.append(mask)
        #print mask
        min_v =  np.ma.min(np.ma.masked_array(m[c], mask))
        pp = np.ma.max(np.ma.masked_array(m[c], mask)) - min_v
        m[c] = (m[c] - min_v) / pp
    return m, mask_a
```

```python
m, mask_a = norm_mask(data)
```
