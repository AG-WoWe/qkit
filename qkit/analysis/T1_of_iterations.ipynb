```python
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

def f_exp(t, Td, a, offs):
    return a*np.exp(-t/Td)+offs
```

Loading the data file can take some seconds, if the file is large

```python
filename = 'D:\\20150331/002225_avg_T1/002225_avg_T1.dat'
data=np.loadtxt(filename)
```

```python
#reshape data such that you have the sweep points along one axis and the iterations along the other
matrix=np.reshape(data[:,3],(data[-1,0]+1,len(data[:,0])/(data[-1,0]+1)))
dates=np.reshape(data[:,4],((data[-1,0]+1,len(data[:,0])/(data[-1,0]+1))))
xaxis = data[:len(data[:,0])/(data[-1,0]+1),1]
```

Show all iterations in one plot

```python
plt.plot(xaxis,matrix.T)
pass
```

Show the mean of all iterations, plot linear and exponential

```python
mean=-np.mean(matrix,axis=0)
plt.figure(figsize=(18,5))
plt.subplot(1,2,1)
plt.plot(xaxis,np.mean(matrix,axis=0))
plt.subplot(1,2,2)
plt.plot(xaxis,mean-mean[-1])
plt.yscale('log')
```

In the following, two things are done: 
* For every single iteration, a T1 time is fitted (shown green in the plot)
* **blocksize** number of iterations are averaged and T1 times are fitted for each average (blue)

```python
blocksize=50
T1 = []
for row in matrix:
     T1=np.append(T1,curve_fit(f_exp, np.linspace(0,200,62),row, (50,0.25,1.5))[0][0]) # expected T1, amplitude, offset
avg=[]
T1avg=[]
for i in range(len(matrix)-blocksize+1):
    row=np.mean(matrix[i:i+blocksize],axis=0)
    avg=np.append(avg,row)
    T1avg=np.append(T1avg,curve_fit(f_exp, np.linspace(0,200,62),row, (50,0.25,1.5))[0][0])

font = {'weight' : 'normal', 'size' : 20, 'family' : 'serif'}
plt.rc('font', **font)
labelsize=22
    
plt.figure(figsize=(20,10))
plt.plot((dates[:,0]-dates[0,0])/3660,T1,'g:')
plt.plot((dates[(blocksize/2):len(matrix)-(blocksize/2)+1,0]-dates[0,0])/3660,T1avg)
plt.xlabel('Hours after start of measurement')
plt.ylabel(r"$T_1 \,(\mu s)$",fontsize=labelsize)
plt.grid()
plt.ylim((np.percentile(T1,.2)*.9,np.percentile(T1,99.9)*1.1)) # This helps to prevent destroying your scale by spikes
plt.savefig(filename[:-4]+'_fit.png',bbox_inches = 'tight',transparent=True)
```
