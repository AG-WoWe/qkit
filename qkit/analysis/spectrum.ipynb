### Normalization and masking

```python
def norm_mask(tolerance=0.5):
    m = m_dat.T
    mask_a = []
    for c in range(m.shape[0]):
        lower_th = np.mean(m[c][np.nonzero(m[c])]) - tolerance*np.abs(np.mean(m[c][np.nonzero(m[c])]))
        upper_th = np.mean(m[c][np.nonzero(m[c])]) + tolerance*np.abs(np.mean(m[c][np.nonzero(m[c])]))
        mask = np.any((m[c]<lower_th,m[c]>upper_th,m[c]==0),axis=0)
        mask_a.append(mask)
        #print mask
        min_v =  np.ma.min(np.ma.masked_array(m[c], mask))
        pp = np.ma.max(np.ma.masked_array(m[c], mask)) - min_v
        m[c] = (m[c] - min_v) / pp
    return m, mask_a
```

```python
m, mask_a = norm_mask()
```

```python
h = 6.626e-34
e = 1.6022e-19
phi0 = h/(2*e)
Ec = 0.214
d = 0.1
Lgeo=0
```

```python
Ic=85e-9
h/(16*np.pi**2*e**2*(Lgeo+phi0/(2*np.pi*Ic)))
```

```python
def f_transmon_disp(I, Ic, Lgeo, dphi, a):
    #return np.sqrt(8*Ec*1e-9*h/(16*np.pi**2*e**2*(Lgeo+phi0/(2*np.pi*Ic*np.abs(np.cos(np.pi*a*I+dphi))))))
    Ictot = Ic*np.abs(np.cos(np.pi*a*I+dphi))*np.sqrt(np.abs(1+(d*np.tan(np.pi*a*I+dphi))**2))
    return np.sqrt(8*Ec*1e-9*h/(16*np.pi**2*e**2*(Lgeo+phi0/(2*np.pi*Ictot))))-Ec
```

```python
popt, pcov = curve_fit(f_transmon_disp, x_val, 1e-9*np.array(f01).squeeze(),p0=[85e-9,0.4e-9,0.85,0.1])
```
