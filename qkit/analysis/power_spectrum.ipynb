```python
import matplotlib.colors as col
import matplotlib.cm as cm
```

```python
stridat = "153252_vna_mw power (dBm)mw frequency (Hz).dat"
data = np.genfromtxt(stridat).T
```

```python
print data.shape
```

```python
row_x = 0
row_y = 1
row_d = 6
```

```python
x_val = np.unique(data[row_x])
len(x_val)
```

```python
y_val = np.extract(data[row_x] == data[row_x][0],data[row_y])
len(y_val)
```

```python
m_dat_arr = []
for y in y_val:
    m_dat_arr.append(np.extract(data[row_y] == y,data[row_d]))
```

```python
m_dat = np.array(m_dat_arr)
m_dat.shape
```

```python
def data_mask(tolerance = 1):   #creates data mask
    m = m_dat.T
    mask_a = []
    for c in range(m.shape[0]):
        lower_th = np.mean(m[c]) - tolerance*np.abs(np.mean(m[c]))
        upper_th = np.mean(m[c]) + tolerance*np.abs(np.mean(m[c]))
        mask_a.append(np.any((m[c] < lower_th, m[c] > upper_th),axis=0))   #create mask array
    return np.array(mask_a).T
```

```python
def normalize(m):   #normalize columnwise
    for c in range(len(m[0])):
        min_v = np.ma.min(np.ma.masked_array(m_dat, data_mask())[:,c])
        m[:,c] = m[:,c] - min_v
        max_v = np.ma.max(np.ma.masked_array(m_dat, data_mask())[:,c])
        m[:,c] /= max_v
    return m
```

```python
m_dat_n = normalize(m = m_dat)
```

```python
X, Y = np.meshgrid(x_val, y_val)
```

```python
startcolor = 'black'  # a dark olive 
midcolor = 'blue'    # a bright yellow
endcolor = 'white'    # medium dark red
c_cmap = col.LinearSegmentedColormap.from_list('olive',[startcolor,midcolor,endcolor])
# extra arguments are N=256, gamma=1.0
cm.register_cmap(cmap=c_cmap)
```

```python
font = {'weight' : 'normal', 'size' : 20}
plt.rc('font', **font)
labelsize=27
```

```python
fig, axes = plt.subplots(figsize=(15,7))
p = axes.pcolor(X, Y, np.ma.masked_array(m_dat_n, data_mask()), cmap=c_cmap)
cb = fig.colorbar(p, ax=axes)
cb.set_label(r'$\langle\hat n\rangle \rightarrow \arg (S_{11})\,\mathrm{(a.u.)}$', fontsize=labelsize)
#p.set_clim(0.7,1)
axes.set_xlim(x_val[0], x_val[-1])
axes.set_ylim(y_val[0], y_val[-1])
axes.set_xlabel(r'$P_{\mu w}\,(\mathrm{dBm})$', fontsize=labelsize)
axes.set_ylabel(r'$f_{\mu w}\,(\mathrm{Hz})$', fontsize=labelsize)
#fig.tight_layout()
```
