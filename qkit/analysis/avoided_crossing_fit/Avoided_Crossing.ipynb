### Avoided crossing module:

```python
import os,sys,time
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import scipy.constants as sc
import scipy.optimize as fit
#2d array analysis
import scipy.ndimage as scd

from qkit.analysis import dat_reader as dr
#os.chdir("d:/qkit")
from qkit.analysis.avoided_crossing_fit import avoided_crossing_fit as acf
```

### Reading Data

```python
#Reading test data
#!!! Enter correct directory of the test file !!!
datatest, fn=dr.load_data("204527_3D_current(mA)_mwfrequency(Hz).h5",['curr', 'mw freq', 'pha'])
curr, freq, phase_raw = datatest

#Reshape phase (h5 load has shape (x, y, 1), but we need (x, y))
phase_raw = phase_raw.reshape(53, 500).T
```

```python
phase = phase_raw
phase[phase > 1.05] = 1.05
phase[phase < 1] = 1


#Applying rank and gaussian filters (this is sometimes nescessary to get good results)
#rank is either -1, 0, 1
phase = scd.filters.rank_filter(phase, rank = 1, footprint = np.ones((5,1)))
phase = scd.filters.gaussian_filter1d(phase.T, sigma=5).T


plt.pcolormesh(curr, freq, phase, cmap="coolwarm", vmin=phase.min(), vmax=phase.max())
plt.colorbar()
```

```python
#Preselect data if nescessary
phase = phase[:, 2:-2]
phase_raw = phase_raw[:, 2:-2]
curr = curr[2:-2]
```

```python
#avoided_crossing_peakdata needs dips, thus add a - if you have dips in magnitude/if the derivative of your phase signal has dips
#choose to fit data only if there are few to no missmatched points
lok_max, guess=acf.avoided_crossing_peakdata(curr, freq, -phase, data_type = 1, mode = "max", save_fig = False)
```

```python
#Check for errors
nz0 = np.nonzero(lok_max[:,0])[0]
nz1 = np.nonzero(lok_max[:,1])[0]
plt.plot(curr[nz0], lok_max[nz0,0])
plt.plot(curr[nz1], lok_max[nz1,1])
```

```python
#Remove errors and swap points if nescessary
swapindex = np.argmax(lok_max[:, 0])
lok_max[swapindex, 0], lok_max[swapindex, 1] =lok_max[swapindex, 1], 0
lok_max[swapindex+1, 1] = 0
nz1 = np.nonzero(lok_max[:,1])[0]

plt.plot(curr[nz0], lok_max[nz0,0])
plt.plot(curr[nz1], lok_max[nz1,1])
#Now it looks way better!
```

```python
#Use crossing_fit, if the initial guess of parameters is poor better leave it empty
fit_params, fit_errors = acf.crossing_fit(curr[nz0], curr[nz1], lok_max[nz0,0], lok_max[nz1,1], guess = guess)
```

```python
font = {'weight' : 'normal', 'size' : 20}
plt.rc('font', **font)
labelsize=27
```

```python
#Create a nice overlay of data and fit:

#Shift current for pcolormesh (looks a lot better)
fig, axes = plt.subplots(figsize=(17,7))
cp = curr - .5*(curr[1]- curr[0])*np.ones(curr.shape)
plt.pcolormesh(cp, freq*10**-9, phase_raw, cmap='bone')
cbar = plt.colorbar()
xlin = np.linspace(cp.min(), cp.max(), 100)
plt.plot(xlin, acf.avoided_crossing(xlin, fit_params, -1), 'r')
plt.plot(xlin, acf.avoided_crossing(xlin, fit_params, 1), 'r')
plt.xlim(cp.min(), cp.max())
plt.ylim(freq.min()*10**-9, freq.max()*10**-9)
#plt.plot(curr[nz0], lok_max[nz0, 0]*10**-9, 'w*', markersize = 5)
#plt.plot(curr[nz1], lok_max[nz1, 1]*10**-9, 'w*', markersize = 5)
plt.xlabel(r'$I_{coil}^{dc}\,(\mathrm{mA})$', fontsize=labelsize)
plt.ylabel(r'$f\,(\mathrm{GHz})$', fontsize=labelsize)
cbar.set_label(r'$\phi\,(\mathrm{a.u.})$', rotation='90', fontsize = labelsize,labelpad=5)
#plt.savefig('my_crossing.png', dpi = 500)
```
