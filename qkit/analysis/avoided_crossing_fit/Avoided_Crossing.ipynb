### Avoided crossing module:

```python
import os, sys, time
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import scipy.constants as sc
import scipy.optimize as fit

import scipy.ndimage as scd

from qkit.analysis import dat_reader as dr
from qkit.analysis.avoided_crossing_fit import avoided_crossing_fit as acf
#os.chdir("d:/qkit")
```

### Reading Data

```python
#Reading data
#!!! Enter correct directory of the test file !!!
data, fn = dr.load_data("data/crossings/160830_3D_current(mA)_mwfrequency(Hz).h5",['curr', 'mw freq', 'pha'])
curr, freq, signal_raw = data
```

```python
#Reshape signal (h5 load has shape (x, y, 1), but we need (x, y))
signal_raw = signal_raw.reshape(signal_raw.shape[0], signal_raw.shape[1]).T
```

```python
signal = signal_raw

#Applying rank and gaussian filters (this is sometimes nescessary to get good results)
#rank is either -1, 0, 1, footprint i.e. shape of the filter can be adjusted
signal = scd.filters.rank_filter(signal, rank = 1, footprint = np.ones((5,1)))
signal = scd.filters.gaussian_filter1d(signal.T, sigma=5).T

fig, axes = plt.subplots(figsize=(16,8))

plt.pcolormesh(curr, freq, signal, cmap="coolwarm", vmin=signal.min(), vmax=signal.max())
plt.xlim(min(curr), max(curr))
plt.ylim(min(freq), max(freq))
plt.colorbar()
```
