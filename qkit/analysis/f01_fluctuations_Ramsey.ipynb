```python
dr.fit_data('dat_import',fit_function='damped_sine',data_c=1,ps=[1,0.5,None,None,None],data=[delay,dat[761]],nfile='data/Ramsey')
```

```python
dt = []
dterr = []
for i in range(len(it)):
    ret = dr.fit_data('dat_import',fit_function='damped_sine',data_c=1,ps=[1.5,0.5,None,None,None],show_plot=False,data=[delay,dat[i]],nfile='data/Ramsey')
    dt.append(ret[0])
    dterr.append(ret[5])
```

```python
mask = np.any((np.array(dt) < -2, np.array(dt) == 1, np.array(dt) > 20, np.array(dterr) > 0.15), axis=0)
```

```python
font = {'weight' : 'normal', 'size' : 20}
plt.rc('font', **font)
labelsize=27
```

```python
fig, axes = plt.subplots(figsize=(15,5))
#axes.errorbar(it,np.ma.masked_array(dt,mask),yerr=dterr)
axes.plot(it/3600*16.854,np.ma.masked_array(dt,mask),'.',color='b',markersize = 5, lw = 0.1, label='T1')
axes.set_xlim(it[0],it[-1]/3600*16.854)
axes.set_ylabel(r'$\Delta f_{01}\,(\mathrm{MHz})$', fontsize=labelsize)
axes.set_xlabel(r'$time\,(\mathrm{h})$', fontsize=labelsize)
fig.tight_layout()
```

```python
fig.savefig('f01_Ramsey.png', dpi=300)
fig.savefig('f01_Ramsey.pdf', dpi=300)
```
