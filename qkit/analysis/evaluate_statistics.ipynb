```python
import os, glob
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
os.chdir('d:\\qkit')
from qkit.gui.notebook import Progress_Bar as pb
from qkit.analysis import dat_reader as dr
```

### specify data directory

```python
data_dir = 'y:\\Cryostats\\Silver_Fridge_UFO\\data'
os.chdir(data_dir)
```

### folder names to be included in the search

```python
foldernames = ['T1T2seq2']#['20150808','20150809']
```

### set file name mask

```python
fmask = 'T1*avg'
```

```python
def get_files(foldernames,fmask):
    files = []
    for folder in foldernames:
        f = glob.glob(folder+'/*/*'+fmask+'.dat')
        f.sort(key=os.path.getmtime)
        files+=f
    os.chdir(data_dir)
    return files
```

```python
files = get_files(foldernames,fmask)
```

```python
files[0]
```

```python
os.chdir(data_dir)
p = pb.Progress_Bar(len(files))
T1 = []
dT1 = []
for f in files:
    rv = dr.fit_data(f,fit_function='exp',show_plot=False)
    T1.append(rv[0])
    dT1.append(rv[3])
    p.iterate()
```

```python
files = get_files(foldernames,'spinecho*avg')
```

```python
os.chdir(data_dir)
p = pb.Progress_Bar(len(files))
T2 = []
dT2 = []
for f in files:
    rv = dr.fit_data(f,fit_function='exp',show_plot=False)
    T2.append(rv[0])
    dT2.append(rv[3])
    p.iterate()
```

```python
import numpy as np
```

```python
xvec = np.linspace(-10.8,-10.5,30)
```

```python
font = {'weight' : 'normal', 'size' : 20}
plt.rc('font', **font)
labelsize=27
```

```python
mask = np.any((np.array(T1) < 1, np.array(T1) > 12, np.array(dT1) > 1), axis=0)
```

```python
mask2 = np.any((np.array(T2) < 2, np.array(T2) > 12, np.array(dT2) > 1), axis=0)
```

```python
fig, axes = plt.subplots(figsize=(15,5))
axes.errorbar(xvec,np.ma.masked_array(T1,mask),yerr=dT1)
axes.plot(xvec,np.ma.masked_array(T1,mask),'*-',color='b',markersize = 11, lw = 0.3, label='T1')
axes.set_xlim(xvec[0],xvec[-1])

axes.errorbar(xvec,np.ma.masked_array(T2,mask2),yerr=dT2)
axes.plot(xvec,np.ma.masked_array(T2,mask2),'*-',color='g',markersize = 11, lw = 0.3, label='T2')
axes.set_xlim(xvec[0],xvec[-1])

axes.legend(loc = 1, prop = dict(size = 15))
```
