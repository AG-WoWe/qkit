```python
from scipy.optimize import curve_fit
```

```python
def stat_sqrt_n(bs, a, offs):
    return offs+a/np.sqrt(bs)
def stat_n_4(bs, a, offs):
    return offs+a/np.power(bs,0.25)
```

```python
data2 = data+data
#y_sl = y_mask[0]
Tstd = []
Trstd = []

bses = range(2,264)
for bs in bses:
    b_mean = []
    for i in range(len(data)):
        b_mean.append(np.mean(data2[i:(i+bs)]))
    Tstd.append(np.std(b_mean))   #std when taking the average in boxes of size bs
    Trstd.append(np.std(b_mean)/np.mean(data))
```

```python
av_time = np.multiply(bses,55./60)
popt2, pcov2 = curve_fit(stat_sqrt_n, av_time, Tstd)
popt4, pcov4 = curve_fit(stat_n_4, av_time, Tstd)
```

```python
fig, axes = plt.subplots(1,2,figsize=(15,5))
axes[0].plot(av_time,Tstd,'*',markersize=9)
axes[0].plot(av_time,stat_sqrt_n(av_time, *popt2))
axes[0].plot(av_time,stat_n_4(av_time, *popt4))
axes[0].set_xlabel(r'$\mathrm{average\,time\,(min)}$', fontsize=labelsize)
axes[0].set_ylabel(r'$\sigma_{T_1}\,(\mathrm{\mu s})$', fontsize=labelsize)
axes[1].plot(av_time,1e3*(Tstd-stat_sqrt_n(av_time, *popt2)), '*', markersize=9)
axes[1].set_xlabel(r'$\mathrm{average\,time\,(min)}$', fontsize=labelsize)
axes[1].set_ylabel(r'$(\sigma_{T_1}-fit)\,(10^{-3}\,\mathrm{\mu s})$', fontsize=labelsize)
#axes[1].set_ylim([-0.02,0.01])
fig.tight_layout()
```
