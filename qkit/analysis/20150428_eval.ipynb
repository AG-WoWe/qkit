```python
data = dr.load_data('182213_vna_mw frequency (Hz).dat')
```

```python
import matplotlib.pylab as plt
```

```python
%matplotlib inline
```

```python
font = {'weight' : 'normal', 'size' : 20}
plt.rc('font', **font)
labelsize=27
```

```python
fig = plt.figure(figsize=(13,8))
plt.plot(data[0][0],data[0][5], 'g')
plt.plot(data[0][0],data[0][5], '*')
plt.xlabel(r'$f_{\mu w}\,(\mathrm{Hz})$', fontsize=labelsize)
plt.ylabel(r'$\arg(S11)\,(\mathrm{a.u.})$', fontsize=labelsize)
plt.xlim(7.2e9,7.6e9)
fig.tight_layout()
```

```python
fig.savefig('f01_f02.png', dpi=300)
```
